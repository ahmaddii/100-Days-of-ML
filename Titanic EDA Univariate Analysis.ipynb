```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("train2.csv")

df.head()
```

```python
# Categorical Data Analysis

# Count Plot

sns.countplot(df["Survived"])

df["Survived"].value_counts().plot(kind="bar")
```

```python
sns.countplot(df["Pclass"])

df["Pclass"].value_counts().plot(kind="bar")
```

```python
sns.countplot(df["Sex"])

# if we want exact plot number

df["Sex"].value_counts()
```

```python
sns.countplot("Embarked")
```

```python
df["Embarked"].value_counts().plot(kind="bar")
```

```python
# Now if you want same thing in % rather then freququncy
```

```python
# Use Piechart

df["Embarked"].value_counts().plot(kind="pie",autopct='%.2f')
```

```python

# Numerical Data is continues try to create histogram

# Histogram 


import matplotlib.pyplot as plt

plt.hist(df["Age"])

```

```python
#b: DistPlot : improvment of Histogram

sns.distplot(df["Age"]) # ye hamaun probability btata ha probability desnity function
```
