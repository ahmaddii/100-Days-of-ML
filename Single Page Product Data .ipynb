<h1>Single Page Product Scraping of Gaming Ecommerce Website</h1>

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python

url = "https://sandbox.oxylabs.io/products?page=1"
headers = {"User-Agent": "Mozilla/5.0"}  # pretend to be a browser

response = requests.get(url,headers=headers)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text,"html.parser")

products = soup.find_all("div",{"class": "product-card"})

print(len(products))
```

```python
first = products[0]
```

```python
title = first.find("h4",{"class":"title"}).get_text(strip=True)
price = first.find("div",{"class":"price-wrapper"}).get_text(strip=True)
```

```python
print(title)
print(price)
```

```python
data = []

for product in products:

    # get 4 fields from the website
    
    title = product.find("h4", {"class": "title"}).get_text(strip=True)
    price = product.find("div", {"class": "price-wrapper"}).get_text(strip=True)

   
    availability_tag = product.find("p", {"class": ["in-stock", "out-of-stock"]})
    availability = availability_tag.get_text(strip=True) if availability_tag else None

    link_tag = product.find("a", href=True)
    link = "https://sandbox.oxylabs.io" + link_tag["href"] if link_tag else None

    data.append({
        "title": title,
        "price": price,
        "availability": availability,
        "link": link
    })

print(len(data))
```

```python
df = pd.DataFrame(data)

df.to_csv("single_page_products.csv",index=False)
```
