```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


headers = {"User-Agent": "Mozilla/5.0"}


all_data = []
total_pages = 94

for page in range(1, total_pages + 1):

    print(f"Scraping page {page}/{total_pages}...")

    url = f"https://sandbox.oxylabs.io/products?page={page}"

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

    except requests.RequestException as e:
        print(f"Page {page} failed: {e}")
        continue

    soup = BeautifulSoup(response.text, "html.parser")
    products = soup.find_all("div", class_="product-card")

    print(f"Found {len(products)} products")

    for product in products:

        title_tag = product.find("h4", class_="title")
        price_tag = product.find("div", class_="price-wrapper")
        availability_tag = product.find("p", class_=["in-stock", "out-of-stock"])
        link_tag = product.find("a", href=True)

        all_data.append({
            "title": title_tag.get_text(strip=True) if title_tag else None,
            "price": price_tag.get_text(strip=True) if price_tag else None,
            "availability": availability_tag.get_text(strip=True) if availability_tag else None,
            "link": "https://sandbox.oxylabs.io" + link_tag["href"] if link_tag else None
        })

df = pd.DataFrame(all_data)

print("\nScraping completed!")
print(f"Total products scraped: {len(df)}")
print(df.shape)

df.head()
```

```python
df.to_csv("products.csv",index=False)
```
