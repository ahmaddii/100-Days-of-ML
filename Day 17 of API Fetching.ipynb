<h1>Loop through the 549 Pages and append them into the dataframe by ignoring the index</h1>

```python
for i in range(1,550):
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=4c30f5936a227aca46fd15ffa436045c&language=en-US&page={i}")
    temp_df = pd.DataFrame(response.json()["results"])[["id","title","overview","release_date","popularity","vote_average","vote_count"]]
    df = pd.concat([df,temp_df],ignore_index=True)
    
```

```python
df.to_csv("movies.csv") #then save them to the csv as df
```
