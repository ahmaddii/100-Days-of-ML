```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

headers = {
    "User-Agent": (
        "Mozilla/5.0 (X11; Linux x86_64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/137.0.0.0 Safari/537.36"
    ),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1",
}

url = "https://pk.indeed.com/jobs?q=data+scientist&l=Pakistan"

response = requests.get(url,headers=headers)

soup = BeautifulSoup(response.text,"lxml") # lxml is built in HTML parser
```

```python
titles = []
companies = []
locations = []
salaries = []
descriptions = []
links = []


# -----------------------------
# Loop Through Jobs
# -----------------------------
for card in cards:

    # ---------- Title ----------
    title = card.find("span", title=True)
    if title:
        title = title["title"]
    else:
        title = "Not Available"

    # ---------- Company ----------
    company = card.find("span", attrs={"data-testid": "company-name"})
    if company:
        company = company.text.strip()
    else:
        company = "Not Available"

    # ---------- Location ----------
    location = card.find("div", attrs={"data-testid": "text-location"})
    if location:
        location = location.text.strip()
    else:
        location = "Not Available"

    # ---------- Salary ----------
    salary = card.find("div", attrs={"data-testid": "attribute_snippet_testid"})
    if salary:
        salary = salary.text.strip()
    else:
        salary = "Not Mentioned"

    # ---------- Job Link ----------
    link = card.find("a", class_="jcs-JobTitle")

    if link:
        job_link = "https://pk.indeed.com" + link["href"]
    else:
        job_link = "Not Available"

    # ---------- Job Description ----------
    job_description = "Not Available"

    if job_link != "Not Available":

        try:
            job_response = requests.get(job_link, headers=headers)

            if job_response.status_code == 200:

                job_soup = BeautifulSoup(job_response.text, "lxml")

                description = job_soup.find(
                    "div",
                    id="jobDescriptionText"
                )

                if description:
                    job_description = description.get_text(" ", strip=True)

        except Exception:
            pass

        # Be polite to the server
        time.sleep(1)

    # ---------- Append ----------
    titles.append(title)
    companies.append(company)
    locations.append(location)
    salaries.append(salary)
    links.append(job_link)
    descriptions.append(job_description)

    df = pd.DataFrame({

        "Title": titles,
        "Company": companies,
        "Locations": locations,
        "Salaries": salaries,
        "links": links,
        "Descriptions": descriptions
        
    }
        
    )

    print(df.head())
    
    
```

```python
df.to_csv("job_listings.csv",index=False)

print("File Created")
```
