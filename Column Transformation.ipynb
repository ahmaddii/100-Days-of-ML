```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("covid_toy.csv")
```

```python
df.head()
```

```python
df.isnull().sum() # fever has 10 missing values so we have to fix it
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=["has_covid"]),df["has_covid"],test_size=0.2)
```

```python
X_train
```

<h1>Mushkil ZindGi</h1>

```python
# adding simple Imputer to fever column

si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[["fever"]])

X_test_fever = si.transform(X_test[["fever"]])
```

```python
X_train_fever.shape
```

```python
X_test_fever.shape
```

```python
oe = OrdinalEncoder(categories=[["Mild","Strong"]])

X_train_cough = oe.fit_transform(X_train[["cough"]])

X_test_cough = oe.transform(X_test[["cough"]])
```

```python
X_train_cough.shape
```

```python
X_test_cough.shape
```

```python
one = OneHotEncoder(drop="first",sparse_output=False)

X_train_gender_city = one.fit_transform(X_train[["gender","city"]])

X_test_gender_city = one.transform(X_test[["gender","city"]])
```

```python
X_train_gender_city.shape
```

```python
X_test_gender_city.shape
```

```python
X_train_Age = X_train.drop(columns=["gender","fever","cough","city"]).values

X_test_Age = X_test.drop(columns=["gender","fever","cough","city"]).values
```

```python
X_train_Age.shape
```

```python
X_train_transformed = np.concatenate((X_train_Age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)

X_test_transformed = np.concatenate((X_test_Age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)
```

<h1>Mentors Zindgi</h1>

```python
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),["fever"]),
    ("tnf2",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
    ("tnf3",OneHotEncoder(drop="first",sparse_output=False),["gender","city"])
],remainder="passthrough") 

```
